```python
def unit_conversion():
    """
    Performs unit conversions for length, weight, and volume.

    The function prompts the user to:
    1. Select the type of conversion (length, weight, volume).
    2. Enter the value to convert.
    3. Choose the specific conversion direction.

    Parameters:
    None

    Returns:
    None
    Displays the converted result or error messages directly.
    """
    try:
        print("Select Conversion Type:")
        print("1. Length")
        print("2. Weight")
        print("3. Volume")

        choice = input("Enter your choice (1/2/3): ")
        value = float(input("Enter the value to convert: "))

        if choice == "1":
            print("\nLength Conversion")
            print("1. Meters (m) → Feet (ft)")
            print("2. Feet (ft) → Meters (m)")

            option = input("Select option (1/2): ")

            if option == "1":
                result = value * 3.28084
                print("Converted value:", result, "ft")

            elif option == "2":
                result = value / 3.28084
                print("Converted value:", result, "m")

            else:
                print("Error: Invalid length conversion option.")

        elif choice == "2":
            print("\nWeight Conversion")
            print("1. Kilograms (kg) → Pounds (lbs)")
            print("2. Pounds (lbs) → Kilograms (kg)")

            option = input("Select option (1/2): ")

            if option == "1":
                result = value * 2.20462
                print("Converted value:", result, "lbs")

            elif option == "2":
                result = value / 2.20462
                print("Converted value:", result, "kg")

            else:
                print("Error: Invalid weight conversion option.")

        elif choice == "3":
            print("\nVolume Conversion")
            print("1. Liters (L) → Gallons (gal)")
            print("2. Gallons (gal) → Liters (L)")

            option = input("Select option (1/2): ")

            if option == "1":
                result = value * 0.264172
                print("Converted value:", result, "gal")

            elif option == "2":
                result = value / 0.264172
                print("Converted value:", result, "L")

            else:
                print("Error: Invalid volume conversion option.")

        else:
            print("Error: Unsupported conversion type.")

    except ValueError:
        print("Error: Please enter a valid numeric value.")
```

```python
unit_conversion()
```

```python
def find_sum(numbers):
    """
    Calculates the sum of a list of numbers.

    Parameters:
    numbers (list): List of numeric values.

    Returns:
    float: Sum of the numbers.
    """
    return sum(numbers)


def find_average(numbers):
    """
    Calculates the average of a list of numbers.

    Parameters:
    numbers (list): List of numeric values.

    Returns:
    float: Average of the numbers.
    """
    return sum(numbers) / len(numbers)


def find_maximum(numbers):
    """
    Finds the maximum number in a list.

    Parameters:
    numbers (list): List of numeric values.

    Returns:
    float: Largest number in the list.
    """
    return max(numbers)


def find_minimum(numbers):
    """
    Finds the minimum number in a list.

    Parameters:
    numbers (list): List of numeric values.

    Returns:
    float: Smallest number in the list.
    """
    return min(numbers)


def calculation():
    """
    Prompts the user to select a mathematical operation and input a list of numbers.
    Performs the selected operation and prints the result.
    Handles invalid input such as non-numeric values or empty lists.
    """

    try:
        print("Select Operation:")
        print("1. Sum")
        print("2. Average")
        print("3. Maximum")
        print("4. Minimum")

        choice = input("Enter your choice (1/2/3/4): ")

        user_input = input("Enter numbers separated by spaces: ")

        numbers = [float(num) for num in user_input.split()]

        if len(numbers) == 0:
            raise ValueError("Empty list")

        if choice == "1":
            print("Sum:", find_sum(numbers))

        elif choice == "2":
            print("Average:", find_average(numbers))

        elif choice == "3":
            print("Maximum:", find_maximum(numbers))

        elif choice == "4":
            print("Minimum:", find_minimum(numbers))

        else:
            print("Invalid operation selected.")

    except ValueError:
        print("Error: Please enter valid numeric values.")
```

```python
calculation()
```

```python
def extract_every_other(lst):
    """
    Extracts every other element from a list starting from the first element.

    Parameters:
    lst (list): The input list.

    Returns:
    list: A new list containing every other element from the original list.
    """
    return lst[::2]

```

```python
extract_every_other([1,2,3,4,5,6])
```

```python
def get_sublist(lst, start, end):
    """
    Returns a sublist from the given list starting from index 'start'
    and ending at index 'end' (inclusive).

    Parameters:
    lst (list): The original list.
    start (int): Starting index.
    end (int): Ending index.

    Returns:
    list: A sublist from start to end (inclusive).
    """
    return lst[start:end+1]
```

```python
numbers = [1, 2, 3, 4, 5, 6]
result = get_sublist(numbers, 2, 4)
print(result)
```

```python
def reverse_list(lst):
    """
    Reverses a list using slicing.

    Parameters:
    lst (list): The input list.

    Returns:
    list: A new list containing the elements of the original list in reverse order.
    """
    return lst[::-1]
```

```python
reverse_list([1,2,3,4,5])
```

```python
def remove_first_last(lst):
    """
    Removes the first and last elements of a list using slicing.

    Parameters:
    lst (list): The input list.

    Returns:
    list: A new list without the first and last elements.
    """
    return lst[1:-1]
```

```python
remove_first_last([1,2,3,4,5])
```

```python
def get_first_n(lst, n):
    """
    Returns the first n elements from a list using slicing.

    Parameters:
    lst (list): The input list.
    n (int): The number of elements to extract from the beginning.

    Returns:
    list: A new list containing the first n elements of the input list.
    """
    return lst[:n]
```

```python
get_first_n([0,7,8,10,22], 2)
```

```python
def reverse_skip(lst):
    """
    Extracts every second element from the list in reverse order,
    starting from the second-to-last element.

    Parameters:
    lst (list): The input list.

    Returns:
    list: A new list containing every second element from the second-to-last element moving backward.
    """
    return lst[-2::-2]
```

```python
A = np.array([[3, 4],
              [7, 8]])
B = np.array([[5, 3],
              [2, 1]])
```

```python
A_inv = np.linalg.inv(A)
I_check = np.dot(A, A_inv)
print("A · A⁻¹ =\n", I_check)
```

```python
AB = np.dot(A, B)
BA = np.dot(B, A)
print("A · B =\n", AB)
print("B · A =\n", BA)
print("Are they equal?", np.array_equal(AB, BA))
```

```python
AB_T = np.dot(A, B).T
B_TA_T = np.dot(B.T, A.T)
print("(AB)ᵀ =\n", AB_T)
print("Bᵀ · Aᵀ =\n", B_TA_T)
print("Are they equal?", np.array_equal(AB_T, B_TA_T))
```

**Experiment: How Fast is Numpy?**

```python
import time
import numpy as np

# 1. Element-wise Addition

size = 1_000_000

# Python lists
list1 = list(range(size))
list2 = list(range(size))

start = time.time()
list_add = [a + b for a, b in zip(list1, list2)]
end = time.time()
print("Python List Addition Time:", end - start, "seconds")

# NumPy arrays
arr1 = np.arange(size)
arr2 = np.arange(size)

start = time.time()
arr_add = arr1 + arr2
end = time.time()
print("NumPy Array Addition Time:", end - start, "seconds")

# 2. Element-wise Multiplication

# Python lists
start = time.time()
list_mul = [a * b for a, b in zip(list1, list2)]
end = time.time()
print("Python List Multiplication Time:", end - start, "seconds")

# NumPy arrays
start = time.time()
arr_mul = arr1 * arr2
end = time.time()
print("NumPy Array Multiplication Time:", end - start, "seconds")

# 3. Dot Product

# Python lists
start = time.time()
dot_list = sum(a * b for a, b in zip(list1, list2))
end = time.time()
print("Python List Dot Product Time:", end - start, "seconds")

# NumPy arrays
start = time.time()
dot_array = np.dot(arr1, arr2)
end = time.time()
print("NumPy Array Dot Product Time:", end - start, "seconds")

# 4. Matrix Multiplication

matrix_size = 1000

# Python lists
mat1 = [[j for j in range(matrix_size)] for i in range(matrix_size)]
mat2 = [[j for j in range(matrix_size)] for i in range(matrix_size)]

start = time.time()
# Manual matrix multiplication using lists
result = [[sum(mat1[i][k] * mat2[k][j] for k in range(matrix_size))
           for j in range(matrix_size)]
          for i in range(matrix_size)]
end = time.time()
print("Python List Matrix Multiplication Time:", end - start, "seconds")

# NumPy arrays
np_mat1 = np.arange(matrix_size*matrix_size).reshape(matrix_size, matrix_size)
np_mat2 = np.arange(matrix_size*matrix_size).reshape(matrix_size, matrix_size)

start = time.time()
np_result = np.dot(np_mat1, np_mat2)
end = time.time()
print("NumPy Array Matrix Multiplication Time:", end - start, "seconds")
```

**Observations**

NumPy is significantly faster than Python lists for all element-wise operations and dot products because:

NumPy uses vectorized operations in C.

Python lists use pure Python loops, which are slow.

Dot product is particularly slow for Python lists because it performs 1,000,000 multiplications and additions in Python, whereas NumPy does it in compiled code.

Matrix multiplication with Python lists of size 1000×1000 is impractical, it will take tens of minutes, while NumPy completes it in less than a second.

Speedup increases with array size — larger arrays make NumPy’s advantage even more visible.

**Conclusion**

For large numerical computations, always use NumPy arrays instead of Python lists.

Python lists are okay for small-scale calculations, but for millions of elements or matrix operations, NumPy is orders of magnitude faster.
